```python
import os
from PIL import Image
import cv2
from tqdm import tqdm
```

```python
input_folder = "Research Dataset/latest"
output_folder = "Research Dataset/latest/augmented"

os.makedirs(output_folder, exist_ok=True)
```

```python
os.makedirs(output_folder, exist_ok=True)

for class_folder in tqdm(os.listdir(input_folder), desc="Processing classes"):
    class_path = os.path.join(input_folder, class_folder)
    if os.path.isdir(class_path):
        output_class_folder = os.path.join(output_folder, class_folder)
        os.makedirs(output_class_folder, exist_ok=True)
        
        for filename in tqdm(os.listdir(class_path), desc=f"Converting {class_folder} images"):
            if filename.endswith(".png"):
                try:
                    img = Image.open(os.path.join(class_path, filename))
                    output_filename = os.path.splitext(filename)[0] + ".jpg"
                    img.save(os.path.join(output_class_folder, output_filename), "JPEG")
                except Exception as e:
                    print(f"Error converting {filename}: {e}")

            img_path = os.path.join(output_class_folder, filename)
            try:
                img = cv2.imread(img_path)
                if img is None:
                    print(f"Error loading image: {img_path}")
                    continue
                img_resized = cv2.resize(img, (224, 224))
                cv2.imwrite(img_path, img_resized)
            except Exception as e:
                print(f"Error processing {filename}: {e}")
```
